## Part 2 
Visualize a single trajectory of the Lorenz system starting at x0 = (10, 10, 10) and (10+1e-8, 10, 10)

```python
dt = 0.01
x1, y1, z1 = cd.get_lorenz_trajectory(dt=dt, t_end = 1000, start=[10, 10, 10], sigma=10, beta=8.0 / 3.0, rho=28)
x2, y2, z2 = cd.get_lorenz_trajectory(dt=dt, t_end = 1000, start=[10+1e-8, 10, 10], sigma=10, beta=8.0 / 3.0, rho=28)

cd.plot_lorenz_trajectory(x1, y1, z1, linewidth = 0.3, save = False, filename='lorenz_1.png')
cd.plot_lorenz_trajectory(x2, y2, z2, linewidth = 0.3, save = False, filename='lorenz_2.png')
```

Plot the difference between the two trajectories over time

```python
# Compute the squared distance between the trajectories and plot it
squared_dist = cd.plot_difference_between_trajectories(x1, y1, z1, x2, y2, z2, save = True, filename='lorenz_diff.png')

# Find the first index where the difference is greater than 1
first_index = np.argmax(squared_dist > 1)
# Print the time (not time step) that the difference is greater than 1
print('The time that the difference is greater than 1 is: ', first_index*dt)
```
