## Bifurcation Diagram for $\dot{x} = \alpha - x^2$

For $\alpha < 0$, the system has no steady states.

For $\alpha = 0$, the system has one steady state at $0$.

For $\alpha > 0$, the system has two steady states:
one saddle point steady state at $x_0 = - \sqrt{\alpha}$
and one stable steady state at $x_1 = \sqrt{\alpha}$.

```python
# Create function for the stable steady state. Note: for negative values, np.nan is returned
def steady_state_stable(alpha):
    return np.sqrt(alpha)

# Create function for the saddle point steady state.
def steady_state_unstable(alpha):
    return -np.sqrt(alpha)

# Plot the bifurcation diagram of the first system for alpha_range
alpha_range = [-2, 2]
plot_title = f"First System"
bifurcation.plot_bifurcation_diagram(steady_state_stable,
                                     steady_state_unstable,
                                     alpha_range,
                                     plot_title=plot_title)
```

## Bifurcation Diagram for $\dot{x} = \alpha - 2x^2 - 3$

For $\alpha < 3$, the system has no steady states.

For $\alpha = 3$, the system has one steady state at $0$.

For $\alpha > 3$, the system has two steady states:
one saddle point steady state at $x_0 = - \sqrt{\frac{\alpha - 3}{2}}$
and one stable steady state at $x_1 = \sqrt{\frac{\alpha - 3}{2}}$.

```python
# Create function for the stable steady state.
def steady_state_stable(alpha):
    return np.sqrt((alpha-3)/2)

# Create function for the saddle point steady state.
def steady_state_unstable(alpha):
    return -np.sqrt((alpha-3)/2)

# Plot the bifurcation diagram of the second system for alpha_range
# Note: The range [1,5] in the second system corresponds to the range [-1,1] in the first system
#  and the range [-1,7] in the second system corresponds to the range [-2,2] in the first system
alpha_range = [-1, 7]
plot_title = f"Second System"
bifurcation.plot_bifurcation_diagram(steady_state_stable,
                                     steady_state_unstable,
                                     alpha_range,
                                     plot_title=plot_title)
```
