Here we reconstructed the different phase images from the book. Sometimes it was sovable using the A matrix. We also calculated the eigenvalues in the process.

```python
eigenvalues = []

# node (stable)
A = np.array([
    [-1, -0.5],
    [-0.5, -1]
])
plot_vectorfield(A, f"node (stable)")
eigenvalues.append(np.linalg.eigvals(A))

# focus (stable)
alpha = 0.1
A = np.array([
    [-alpha, -alpha],
    [0.25, 0]
])
plot_vectorfield(A, f"focus (stable), $\\alpha={alpha}$")
eigenvalues.append(np.linalg.eigvals(A))

# saddle (unstable)
alpha = -0.5
A = np.array([
    [-alpha, -alpha],
    [0.25, 0]
])
plot_vectorfield(A, f"saddle (unstable), $\\alpha={alpha}$")
eigenvalues.append(np.linalg.eigvals(A))

# node (unstable)
A = np.array([
    [1, 0.5],
    [0.5, 1]
])
plot_vectorfield(A, f"node (unstable)")
eigenvalues.append(np.linalg.eigvals(A))

# focus (unstable)
A = np.array([
    [0.1, -1],
    [1, 0.1]
])
plot_vectorfield(A, f"focus (unstable)")
eigenvalues.append(np.linalg.eigvals(A))

# Print the eigenvalues
for i, eigenvalue in enumerate(eigenvalues):
    print(f"Eigenvalues for set {i+1}: {eigenvalue}")
```

The plotting of the eigenvalues corresponding to the plots above. The order of the eigenvalues is the same as the one of the phase images above.

```python
plot_eigenvalues(eigenvalues)
```
