# First Part: Approximate a linear function 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import lstsq
from func_approximation import *
```

```python
# Load the data
data_A = np.loadtxt('linear_function_data.txt')
X_A = data_A[:, 0]
f_A = data_A[:, 1]

# Reshape X_A to be a 2D array
X_A = X_A.reshape(-1, 1)

# Perform least squares linear regression with a regularization parameter
A, _, _, _ = lstsq(X_A, f_A, 1)

# Predict the function values
f_A_pred = X_A @ A

# Plot the original data and the approximated function
plot_function_data(X_A, f_A, f_A_pred, 'Approximation of the Function in Dataset (A) with a Linear Function (Regularized)', 'Approximated Function')
```

# Second Part: Approximate a nonlinear function 

```python
# Load the data
data_B = np.loadtxt('nonlinear_function_data.txt')
X_B = data_B[:, 0]
f_B = data_B[:, 1]

# Reshape X_B to be a 2D array
X_B = X_B.reshape(-1, 1)

# Perform least squares linear regression with a regularization parameter
A, _, _, _ = lstsq(X_B, f_B, 1)

# Predict the function values
f_B_pred = X_B @ A

# Plot the original data and the approximated function
plot_function_data(X_B, f_B, f_B_pred, 'Approximation of the Function in Dataset (B) with a Linear Function', 'Approximated Function')
```

# Third Part: Approximate a nonlinear function with a combination of radial funcitons

```python
# Define the centers and epsilon
centers = np.linspace(-3, 4, 15)
epsilon = 0.9

# Perform the radial basis function approximation
C, Phi = radial_basis_function_approximation(X_B, f_B, centers, epsilon, reg=100)

# Predict the function values
f_B_pred = Phi @ C

# Plot the original data and the approximated function
plot_function_data(X_B, f_B, f_B_pred, 'Approximation of the Function in Dataset (B) with Radial Basis Functions', 'Approximated Function (epsilon=0.8, centers=15)')
```

## Analyze approximation behavior with different values for L and epsilon

```python
# Define the L values and epsilon values
L_values = [4, 8, 15, 25, 50]
epsilon_values = [0.01, 0.1, 0.5, 1, 2]
centers_range = [-3, 4]

# Plot the approximations for the different L and epsilon values
plot_approximations(X_B, f_B, L_values, epsilon_values, centers_range)
```
