## PART 1

```python
# np.set_printoptions(suppress=True)
# ds = np.loadtxt('../Moodle/MI_timesteps.txt', skiprows=1)
# ds = ds[1000:, 1:]

# Read the dataset
data = np.loadtxt('../Moodle/MI_timesteps.txt', delimiter=' ', skiprows=1)

# Remove the burn-in period
data = data[1000:, 1:]

# Select the columns for the delay embedding
# delay_columns = ['areaDensityCounting-PID5', 'areaDensityCounting-PID6', 'areaDensityCounting-PID7']
delay_columns = [0, 1, 2]  # Assuming columns 1, 2, and 3 are selected (The timestep column was sliced)
delay_data = data[:, delay_columns]

# Create delay embeddings
window_size = 351
delay_embeddings = np.lib.stride_tricks.sliding_window_view(delay_data, window_shape=(window_size, 3))
delay_embeddings = delay_embeddings.reshape(-1, window_size * 3)

# Apply PCA with the selected number of components
pca = PCA(n_components=3)
transformed_data = pca.fit_transform(delay_embeddings)

print("Original data shape:", data.shape)
print("Deay embedded data shape:", delay_embeddings.shape)
print("Transformed data shape:", transformed_data.shape)
# The transformed_data now represents the new state space using the principal components
# print(data, data.shape)
# print(delay_embeddings, delay_embeddings.shape)
# print(transformed_data , transformed_data .shape)
```

```python
# Create a 3x3 grid of subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 15), subplot_kw={'projection': '3d'})

# Create 3D scatter plots for each measurement
for i, ax in enumerate(axs.flat):
    ax.scatter(*transformed_data.T, c=data[:transformed_data.shape[0], i], s=1)
    ax.set_title(f'Measurement area {i+1}')
```

## PART 3

```python
# Compute the arclength of the curve in the PCA space
arclengths = np.cumsum(np.sqrt(np.sum(np.diff(transformed_data, axis=0)**2, axis=1)))

# Calculate the change of arclength over time
time_steps = np.arange(1000, len(transformed_data) + 1000)[1:]

# Create a vector field on the arclength
vector_field = np.gradient(arclengths)

print("Arclengths data shape:", arclengths.shape)
print("Vector_fields data shape:", vector_field.shape)

# Plot the vector field
plt.plot(vector_field)
plt.xlabel("Arclength")
plt.ylabel("Velocity on Arclength")
plt.title("Vector Field on Arclength")
plt.show()
```

```python
# Period point search
period_length = np.argmin(np.linalg.norm(transformed_data[20:] - transformed_data[0], axis=1))
periodic_index = period_length + 20

# Print the periodic index
print("Periodic Index:", periodic_index)

period = vector_field[:periodic_index]
vector_field_period = [i * 2 * np.pi / len(period) for i in range(len(period))]

# Plot the vector field
plt.plot(vector_field_period, period)
plt.xlim(0, 2 * np.pi)  # Set x-axis limits to show one period from 0 to 2*pi
plt.xticks([0, 2 * np.pi], ['0', '2'])
plt.xlabel("Arclength")
plt.ylabel("Derivative of Arclength")
plt.title("Vector Field on Arclength")
plt.show()
```
