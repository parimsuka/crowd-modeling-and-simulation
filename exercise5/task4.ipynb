```python
import numpy as np
import matplotlib.pyplot as plt
from task4_utils import *
```

```python
data = get_manifold_data(data_path="takens_1.txt")
# Get the line number in the manifold dataset
time = range(0, len(data[:, 0]))

# Plot the first coordinate against the line number
plot_2d(x=time, y=data[:, 0], save=True, figure_name = 'first_against_line.png', title="First coordinate against line number", xlabel="Line number", ylabel="First coordinate")
```

```python
delta_n = [10, 20, 50, 100] # Delay to be added

# Plot the first coordinate against the first coordinate with delay = 10, 20, 50, 100
for i in delta_n:
    x, y, z = add_delay(data, delta_n=i)
    figure_name_2d = f"first_against_delay_2d_{i}.png"
    plot_2d(x[:, 0], y[:, 0], save=True, figure_name =  figure_name_2d, title=f"First coordinate against first coordinate with delay = {i}", xlabel=r"$x(t)$", ylabel=r"$x(t+\Delta n)$")
    figure_name_3d = f"first_against_delay_3d_{i}.png"
    plot_3d(x[:, 0], y[:, 0], z[:, 0], save=True, figure_name = figure_name_3d, title="", xlabel=r"$x(t)$", ylabel=r"$x(t+\Delta n)$", zlabel=r"$x(t+2\Delta n)$")
```

```python
dt = 0.01
x1, y1, z1 = get_lorenz_trajectory(dt=dt, t_end = 100, start=[10, 10, 10], sigma=10, beta=8.0 / 3.0, rho=28)
plot_lorenz_trajectory(x1, y1, z1, linewidth = 0.7, save = True, filename='lorenz.png')
```

```python
delta_n = [5, 10, 20] # Delay to be added
for i in delta_n:
    # Add delay
    x, y, z = add_delay(x1, delta_n=i)
    # plot the approximation of the Lorenz attractor using x1=x(t), x2=x(t + delta_t), x3=x(t + 2*delta_t)
    filename = f"lorenz_delay_x_{i}.png"
    plot_lorenz_trajectory(x, y, z, linewidth = 0.7, save = True, filename=filename)
```

```python
for i in delta_n:
    # Add delay
    x, y, z = add_delay(z1, delta_n=i)
    # plot the approximation of the Lorenz attractor using x1=x(t), x2=x(t + delta_t), x3=x(t + 2*delta_t)
    filename = f"lorenz_delay_z_{i}.png"
    plot_lorenz_trajectory(x, y, z, linewidth = 0.7, save = True, filename=filename)
```
